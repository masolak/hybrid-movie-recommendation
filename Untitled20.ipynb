```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
```

```python
plot_scaled=(scaler.fit_transform(plot_sim.T)).T
```

```python
pearson_scaled=(scaler.fit_transform(pearson_sim.T)).T
```

To test the two similarity scores independently I choose some movies I've watched before and observe the results. First movie I choose is "Girl, Interrupted", since it's kinda have a theme I think it would be good choice to test the doc2vec model' results. The story of the movie takes place in a mental instution. Some of the  most similar movies according to the doc2vec model have related story to mental ilness or hospital etc but some are not. Tbh I am kinda dissappointed by top2vec's performance. I trained the model 3-4 times with different hyperparamaters, also one time with the full plot dataset that consists of movies that are not seen in the movielens data, too but even that did not improve the result significantly, even sometimes lowered the performance contrast to longer training time and bigger dataset. Top2vec takes serious time to train but I am doubtful towards that it's worth the time. I think simple tf-dt calculation requires one or two minute calculation would not give significantly worse results than doc2vec. But when I mix the results with pearson similarity matrix I can  say I am satisfied with the performance. As for the pearson similarity matrix, the only flaw I can think of that might be it is kinda biased to popular movies, especially when you ask a recommendation for a popular movie. But again I think pearson similarity mesuare balance each other out with the doc2vec results in that way, too.

```python
k=plot_scaled[1778]
k_sort=sorted(k,reverse=True)
for value in k_sort[:20]:
    ind=np.where(k==value)
    print(df.iloc[ind].Title)
```

```python
k=pearson_scaled[1778]
k_sort=sorted(k,reverse=True)
for value in k_sort[:10]:
    ind=np.where(k==value)
    print(df.iloc[ind].Title)
```

```python
def get_recommendations(ids):
    scores=np.zeros((16002))
    for i in ids:
        a=np.add(plot_scaled[i],pearson_scaled[i])
        scores=np.add(scores,a)
    scores_sort=sorted(scores,reverse=True)
    for value in scores_sort[:20]:
        ind=np.where(scores==value)
        print(df.iloc[ind].Title)
```

```python
get_recommendations([1778])
```
